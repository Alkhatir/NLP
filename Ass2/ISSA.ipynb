### Similarity metric

```python
def cosine_similarity(v1, v2):
    return sum([a*b for a, b in zip(v1, v2)]) / (sum([a**2 for a in v1])**0.5 * sum([a**2 for a in v2])**0.5)

similarities = {}
for source_word in source_words:
    similarities[source_word] = {}
    for target_word in target_words[source_word]:
        similarities[source_word][target_word] = cosine_similarity(model[source_word], model[target_word])
print(similarities)
```

### KNN implementation

```python
def knn(source_word, set_words, k=3):
    distances = {}
    for target_word in set_words:
        distances[target_word] = cosine_similarity(model[source_word], model[target_word])
    return sorted(distances, key=distances.get, reverse=True)[:k], sorted(distances.values())[:k]
```

In Assignment 1, the representation of each document was created using a bag of words representation followed by dimensionality reduction. In this task, the document representations are created from the pre-trained word embeddings.

**Map word embeddings to dictionary words (5 points).** For every word in the dictionary (as discussed and created in Assignment 1), fetch the corresponding word embedding from the pre-trained model. If no embedding is found, initialize the corresponding word embedding randomly.

**Document embedding as the average of word embeddings (5 points).** Using the word embeddings, the representation of each document is defined as the *mean of the vectors of each document's words*. In particular, given the document $d$, consisting of words $\left[ v_1, v_2, ..., v_{|d|} \right]$, the document representation $\mathbf{e}_d$ is defined as:

$\mathbf{e}_d = \frac{1}{|d|}\sum_{i=1}^{|d|}{\mathbf{e}_{v_i}}$

where $\mathbf{e}_{v}$ is the vector of the word $v$, and $|d|$ is the length of the document.

**Classification and evaluation (5 points)** Using these new document representations, apply <ins>three classification algorithms</ins> and report the evaluation results (based on accuracy metric) on the test set.



```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import spacy
from collections import Counter
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
# from sent2vec.vectorizer import Vectorizer
```

code from previous assignment to clean, normalise, build vocab and replace oov

```python
data_path = "nlp2023_24_data/"

train_data = pd.read_csv(f"{data_path}thedeep.subset.train.txt", sep=",", names=["sentence_id", "text", "label"])
test_data = pd.read_csv(f"{data_path}thedeep.subset.test.txt", sep=",", names=["sentence_id", "text", "label"])

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
nlp = spacy.load("en_core_web_sm")

def clean_and_normalize_text(text, use_stemming=False, remove_stopwords=True):
    text = text.lower()
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\d+", "NUMTOKEN", text)
    text = re.sub(r"\b(january|february|march|april|may|june|july|august|september|october|november|december)\b", "DATETOKEN", text, flags=re.IGNORECASE)
    tokens = text.split()
    if remove_stopwords:
        tokens = [word for word in tokens if word not in stop_words]
    if use_stemming:
        tokens = [stemmer.stem(word) for word in tokens]  # Stemming
    else:
        tokens = [token.lemma_ if token.text not in ["NUMTOKEN", "DATETOKEN"] else token.text for token in nlp(" ".join(tokens))]
    cleaned_text = " ".join(tokens).replace("NUMTOKEN", "<num>").replace("DATETOKEN", "<date>")
    
    return cleaned_text


def build_vocabulary(data, min_frequency=2):
    all_tokens = " ".join(data['cleaned_text']).split()
    token_counts = Counter(all_tokens) 
    vocabulary = {token for token, count in token_counts.items() if count >= min_frequency}
    vocabulary.add("<OOV>")
    
    return vocabulary, token_counts
    

def replace_oov_tokens(text, vocabulary):
    # Replaces words that are not in the vocab with <OOV>
    tokens = text.split()
    tokens = [token if token in vocabulary else "<OOV>" for token in tokens]
    return " ".join(tokens)



train_data['cleaned_text'] = train_data['text'].apply(clean_and_normalize_text)
test_data['cleaned_text'] = test_data['text'].apply(clean_and_normalize_text)
vocabulary, token_counts = build_vocabulary(train_data, min_frequency=2)
train_data['processed_text'] = train_data['cleaned_text']
test_data['processed_text'] = test_data['cleaned_text'].apply(lambda x: replace_oov_tokens(x, vocabulary))
```

Loading a word embedding model Task A.1

```python
from gensim.models import KeyedVectors
from gensim.downloader import load

word2vec_model = load('word2vec-google-news-300')
```

```python
def random_embedding(dim, seed=42):
    np.random.seed(seed)
    return np.random.uniform(-0.25, 0.25, dim)

def map_vocab_to_embeddings(vocabulary, word2vec_model, embedding_dim=300):
    word_embeddings = {}
    found = 0
    not_found = 0
    for word in vocabulary:
        if word in word2vec_model:
            found += 1
            word_embeddings[word] = word2vec_model[word]
        else:
            not_found += 1
            word_embeddings[word] = random_embedding(embedding_dim)
    print("vocab found:", found)
    print("vocab not found:", not_found)
    return word_embeddings


embedding_dim = word2vec_model.vector_size
word_embeddings = map_vocab_to_embeddings(vocabulary, word2vec_model, embedding_dim)

print(f"Vocab size: {len(vocabulary)}")
```

```python
def compute_document_embedding(doc, word_embeddings, embedding_dim):
    
    tokens = doc.split()
    token_vectors = [word_embeddings[token] for token in tokens if token in word_embeddings]
    
    if token_vectors:
        return np.mean(token_vectors, axis=0)
    else:
        return np.random.uniform(-0.25, 0.25, embedding_dim)


embedding_dim = word2vec_model.vector_size
train_embeddings = np.array([
    compute_document_embedding(doc, word_embeddings, embedding_dim) for doc in train_data['processed_text']
])
test_embeddings = np.array([
    compute_document_embedding(doc, word_embeddings, embedding_dim) for doc in test_data['processed_text']
])

print("Train Document Embeddings Shape:", train_embeddings.shape)
print("Test Document Embeddings Shape:", test_embeddings.shape)
```

```python
class RandomForestTextClassifier:
    def __init__(self, n_estimators=100, max_depth=None):
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.model = RandomForestClassifier(n_estimators=self.n_estimators, max_depth=self.max_depth, random_state=4)
        
    def fit(self, X_train, y_train):
        self.model.fit(X_train, y_train)
        
    def predict(self, X_test):
        return self.model.predict(X_test)
        
    def evaluate(self, X_test, y_test, is_test=False):
        y_pred = self.predict(X_test)
        if is_test:
            print(f"Random Forest Classifier with {self.n_estimators} Trees and {self.max_depth} depth :")
            print("Accuracy:", accuracy_score(y_test, y_pred))
        return accuracy_score(y_test, y_pred)

rf_classifier = RandomForestTextClassifier(n_estimators=100, max_depth=20)
rf_classifier.fit(train_embeddings, train_data['label'])

rf_accuracy = rf_classifier.evaluate(test_embeddings, test_data['label'], is_test=True)
```

```python
class MLPTextClassifier:
    def __init__(self, hidden_layer_sizes=(100,) ,warm_start=False, max_iter=20):
        self.hidden_layer_sizes = hidden_layer_sizes
        self.max_iter = max_iter
        self.model = MLPClassifier(hidden_layer_sizes=self.hidden_layer_sizes, max_iter=self.max_iter, random_state=4, warm_start=warm_start)
        
    def fit(self, X_train, y_train):
        self.model.fit(X_train, y_train)
        
    def predict(self, X_test):
        return self.model.predict(X_test)
        
    def evaluate(self, X_test, y_test, is_test=False):
        y_pred = self.predict(X_test)
        if is_test:
            print(f"MLP Classifier with {self.hidden_layer_sizes} neurons in the hidden layer and {self.max_iter} epochs:")
            print("Accuracy:", accuracy_score(y_test, y_pred))
            print("\nClassification Report:\n", classification_report(y_test, y_pred))
        return accuracy_score(y_test, y_pred)


mlp_classifier = MLPTextClassifier(hidden_layer_sizes=(100,), max_iter=20)
mlp_classifier.fit(train_embeddings, train_data['label'])

mlp_accuracy = mlp_classifier.evaluate(test_embeddings, test_data['label'], is_test=True)
```

```python
class LogisticRegressionClassifier:
    def __init__(self, max_iter=1000):
        self.max_iter = max_iter
        self.model = LogisticRegression(multi_class='ovr', max_iter=self.max_iter, random_state=42)
        
    def fit(self, X_train, y_train):
        self.model.fit(X_train, y_train)
        
    def predict(self, X_test):
        return self.model.predict(X_test)
        
    def evaluate(self, X_test, y_test, is_test=False):
        y_pred = self.predict(X_test)
        if is_test:
            print(f"Logistic Regression Classifier with {self.max_iter} max iterations:")
            print("Accuracy:", accuracy_score(y_test, y_pred))
        return accuracy_score(y_test, y_pred)

logistic_classifier = LogisticRegressionClassifier(max_iter=1000)
logistic_classifier.fit(train_embeddings, train_data['label'])

logistic_accuracy = logistic_classifier.evaluate(test_embeddings, test_data['label'], is_test=True)
```
